### Importar NumPy y cargar vectores

```python
import numpy as np
```

```python
gato = np.array([0.9, 0.8, 0.2])
perro = np.array([0.8, 0.85, 0.15])
computadora = np.array([0.1, 0.3, 0.9])
```

## Dot product

El Producto Dot (también conocido como producto escalar) es una operación matemática que mide la similitud entre dos vectores. Se utiliza comúnmente en álgebra lineal y análisis de datos. El producto punto de dos vectores se calcula multiplicando elemento por elemento y sumando los resultados. La fórmula para el Producto Dot de dos vectores **A** y **B** se expresa como:

$$
\text{Producto Dot}(\mathbf{A}, \mathbf{B}) = \sum_{i=1}^{n} A_i \cdot B_i
$$

Donde:
- **A** y **B** son vectores de longitud \(n\).
- $(A_i)$ y $(B_i)$ son los elementos correspondientes de los vectores **A** y **B**.


```python
dot_product_gato_perro = np.dot(gato,perro)
dot_product_gato_computadora = np.dot(gato, computadora)

print('Producto punto gato-perro: ', dot_product_gato_perro )
print('Producto punto gato-computadora: ',dot_product_gato_computadora )
```

## Euclidean distance

La distancia Euclidiana mide la distancia "directa" o "en línea recta" entre dos puntos en un espacio multidimensional. Se utiliza para calcular la longitud del segmento de línea que conecta dos puntos en un espacio euclidiano. La fórmula para la distancia Euclidiana entre dos puntos $(P)$ y $(Q)$ en un espacio $(n)$-dimensional se expresa como:

$$
\text{Distancia Euclidiana}(P, Q) = \sqrt{\sum_{i=1}^{n} (P_i - Q_i)^2}
$$

Donde:
- $(P_i)$ y $(Q_i)$ son las coordenadas de los puntos $(P)$ y $(Q)$ en la dimensión $(i)$.
- La raíz cuadrada $(\sqrt{\cdot})$ se toma para obtener la distancia en unidades de la dimensión original.

```python
euclidean_distance = np.linalg.norm(gato-perro)

print(euclidean_distance)
```

```python
euclidean_distance = np.linalg.norm(gato-computadora)

print(euclidean_distance)
```

## Cosine similarity

La similitud del coseno es una medida de similitud entre dos vectores en un espacio vectorial. Se utiliza comúnmente en minería de texto, recuperación de información y recomendación. La similitud del coseno se calcula como el coseno del ángulo entre dos vectores. La fórmula para la similitud del coseno entre dos vectores **A** y **B** se expresa como:

$$
\text{Similitud del Coseno}(\mathbf{A}, \mathbf{B}) = \frac{\mathbf{A} \cdot \mathbf{B}}{\|\mathbf{A}\| \cdot \|\mathbf{B}\|}
$$

Donde:
- $(\mathbf{A} \cdot \mathbf{B})$ es el Producto Dot de los vectores **A** y **B**.
- $(\|\mathbf{A}\|)$ y $(\|\mathbf{B}\|)$ son las normas (longitudes) de los vectores **A** y **B**, respectivamente.

```python
dot_product = np.dot(gato, perro)

norm_gato = np.linalg.norm(gato)
norm_perro = np.linalg.norm(perro)

cosine_similarity = dot_product / (norm_gato * norm_perro)
print(cosine_similarity)
```

```python
dot_product = np.dot(gato, computadora)

norm_gato = np.linalg.norm(gato)
norm_computadora = np.linalg.norm(computadora)

cosine_similarity = dot_product / (norm_gato * norm_computadora)
print(cosine_similarity)
```

## Vectores

Podemos definir todas las magnitudes vectoriales mediante listas de python (organizadas en un diccionario de palabras) y producir, a partir de allí, un diccionario de los respectivos vectores (como arreglos de numpy)

```python
# Dimensionalidad de las palabras
valores = {
    'reina':  [9, 7, 8],
    'rey':    [1, 7, 8],
    'mujer':  [9, 7, 0],
    'hombre': [1, 7, 0],
    'niña':   [9, 2, 0],
    'niño':   [1, 2, 0],
}

# Palabras representadas como vectores
vector = { palabra: np.array(magnitudes) \
            for palabra, magnitudes in valores.items() }

for palabra in vector:
    print(f'{palabra:6}: {vector[palabra]}')
```

## Funciones

Podemos crear 2 funciones para calcular fácilmente la distancia y la similitud entre palabras

```python
# Distancia entre palabras (usando distancia euclidiana)
def distancia(palabra1: str, palabra2: str) -> float:
    """
    Calcula la distancia euclidiana entre los vectores
    que representan dos palabras.
    A mayor distancia, menor similitud.
    """
    diff = vector[palabra2] - vector[palabra1]
    return np.linalg.norm(diff)
```

```python
# Similitud entre palabras (usando similitud coseno)
def similitud(palabra1: str, palabra2: str) -> float:
    """
    Calcula la similitud (por coseno) entre los vectores
    que representan dos palabras.
    Valores cercanos a 1 indican mayor similitud.
    """
    vectores = [vector[palabra1], vector[palabra2]]
    dot_product = np.dot(*vectores)
    norm_1, norm_2 = np.linalg.norm(vectores, axis=1)
    return dot_product / (norm_1 * norm_2)
```

## Cálculos de distancia (y similitud)

Ahora podemos usar estas funciones para calcular directamente la distancia (o bien la similitud) entre cualesquiera 2 palabras:

```python
print(distancia('reina', 'mujer'))
```

```python
print(distancia('reina', 'niña'))
```

```python
print(similitud('reina', 'niña'))
```

Comparemos entonces todos los pares de palabras, mediante una iteración:

```python
# palabras ordenadas en una lista de tuplas
pares = [
    ('rey', 'hombre'),
    ('hombre', 'niño'),
    ('reina', 'mujer'),
    ('mujer', 'niña'),
    ('rey', 'reina'),
    ('hombre', 'mujer'),
]

# a modo de encabezado:
print(f"{'Palabras':16} Distancia - Similitud")

# por cada par de palabras, imprimimos las palabras y su distancia y similitud
for par in pares:
    print(f'{par[0]:6} ←→ {par[1]:6}',
          f'{distancia(*par):^9}',
          f'{similitud(*par):10.5f}')
```
